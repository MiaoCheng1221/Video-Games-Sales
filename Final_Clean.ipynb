```python
import seaborn as sns
import math

str_list = [] # empty list to contain columns with strings (words)
for colname, colvalue in New_video.iteritems():
    if type(colvalue[2]) == str:
         str_list.append(colname)
# Get to the numeric columns by inversion            
num_list = New_video.columns.difference(str_list) 
# Create Dataframe containing only numerical features
video_num = New_video[num_list]
f, ax = plt.subplots(figsize=(12, 10))
plt.title('Pearson Correlation of Video Game Numerical Features')
# Draw the heatmap using seaborn
a=video_num.astype(float).corr()
sns.heatmap(a,annot=True)
plt.show()
```
